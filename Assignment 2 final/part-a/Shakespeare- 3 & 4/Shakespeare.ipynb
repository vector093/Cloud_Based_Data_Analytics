```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
import re
```

```
sp = SparkSession.builder.appName("Demo").getOrCreate()
```

```
textlist = spark.sparkContext.textFile("/FileStore/tables/shakespeare_1.txt")
```

```
textlist= textlist.map(lambda x: re.sub("[^0-9a-zA-Z']+", ' ', x).rstrip())
```

```
textlist.collect()
```

```
textlist=textlist.flatMap(lambda x: x.split())
```

```
textlist.count()
```

```
word=textlist.map(lambda x: (x,1))
```

```
word_count=word.reduceByKey(lambda x,y: x+y)
```

```
word_count.collect()
```

```
print("number of times SHAKESPEARE used")
print(word_count.lookup("SHAKESPEARE"))
print("number of times GUTENBERG used")
print(word_count.lookup("GUTENBERG"))
print("number of times WILLIAM used")
print(word_count.lookup("WILLIAM"))
print("number of times LIBRARY used")
print(word_count.lookup("LIBRARY"))
print("number of times COLLEGE used")
print(word_count.lookup("COLLEGE"))
print("number of times WORLD used")
print(word_count.lookup("WORLD"))
print("number of times THIS used")
print(word_count.lookup("THIS"))
```

```
bottomword=word_count.takeOrdered(20, key=lambda x: x[1])
```

```
print("the bottom 20 words in the file are-")
print(bottomword)
```

```
topword=word_count.takeOrdered(20, key=lambda x: -x[1])
```

```
print("the top 20 words in the file are-")
print(topword)
```

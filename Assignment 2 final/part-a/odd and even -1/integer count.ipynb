```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
sp = SparkSession.builder.appName("Demo").getOrCreate()
```

```
firstRDD = sp.sparkContext.textFile("/FileStore/tables/integer.txt")
```

```
firstRDD.count()
```

```
integer=firstRDD.map(lambda x: int(x))
```

```
even=integer.filter(lambda x: x % 2 == 0)
odd= integer.filter(lambda x: x % 2 == 1)
```

```
print("the number of even numbers are-")
even.count()
```

```
print("the number of odd numbers are-")
odd.count()
```

```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
from operator import add
```

```
sp = SparkSession.builder.appName("Demo").getOrCreate()
```

```
salary = sp.sparkContext.textFile("/FileStore/tables/salary.txt")
```

```
salary.count()
```

```
salary=salary.map(lambda x: x.split(" "))
```

```
salary=salary.map(lambda x: (x[0],float(x[1])))
```

```
salary.collect()
```

```
final_salary=salary.reduceByKey(add)
```

```
final_salary.collect()
```
